```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data  
```

```python
wcss = []
silhouette_scores = []
db_scores = []
```

```python
cluster_range = range(2, 11)
```

```python
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    
    #
    silhouette_avg = silhouette_score(X, cluster_labels)
    db_index = davies_bouldin_score(X, cluster_labels)
    wcss_value = kmeans.inertia_  
    
    silhouette_scores.append(silhouette_avg)
    db_scores.append(db_index)
    wcss.append(wcss_value)
```

```python
plt.figure(figsize=(16, 6))


plt.subplot(1, 3, 1)
plt.plot(cluster_range, silhouette_scores, marker='o', linestyle='--', color='g')
plt.title('Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
```

```python

plt.subplot(1, 3, 2)
plt.plot(cluster_range, db_scores, marker='o', linestyle='--', color='b')
plt.title('Davies-Bouldin Index vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
```

```python
plt.subplot(1, 3, 3)
plt.plot(cluster_range, wcss, marker='o', linestyle='--', color='r')
plt.title('WCSS vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')

plt.show()
```
